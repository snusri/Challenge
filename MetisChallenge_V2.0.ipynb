The following notebook presents the code that was used to extract the D3 github repo commit history for the past year of 2016.

First the commit history was obtained from the following git bash commands:

*git clone https://github.com/d3/d3*

*cd ~../d3*

*git log --since=2016-01-01, --before=2017-01-01 \
   --pretty=format:'{%n  "commit": "%H",%n  "author": "%aN <%aE>",%n  "date": "%ad",%n  "message":* *"%f"%n},' *\ $@* | *\*
   *perl* -pe 'BEGIN{print "["}; END{print "]\n"}' | \*
   *perl -pe 's/},]/}]/' > commithistory.json*
   
   


Once the commit history was obtained pandas was used to tranform and aggregate the data accordingly.

```python
import pandas as pd
```

```python
commithist = pd.read_json("commithistory.json")
commithist.head()
```

```python
commithist['date'] = pd.to_datetime(commithist['date'])

commithist["date"].dtype
```

```python
sortedCommit = commithist.groupby(['author', pd.Grouper(key='date', freq='W-MON')])['commit'].sum().reset_index().sort_values('date')
print (sortedCommit)
```

```python
sortedCommit['TotalCommitsInTheWeek'] = sortedCommit.index
```

```python
sortedCommit.reset_index(level = 0, inplace=True)
del sortedCommit["index"]
sortedCommit
```

```python
g = sortedCommit.groupby("date").sum().reset_index()
g.max()
```

Hence from the EDA, most number of commits were made on the week of 2016-11-28. These were cumulatively 99.
